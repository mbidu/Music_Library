```python
song2 = {}

with open(r'C:\Users\mackt\Python\Music Library\Data\Music Collection (Jul 09 21).csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        song2['song{0}'.format(row['ID'])] = {}
        # print(row['Rating'])
        if row['Rating'] == '':
            song2['song{0}'.format(row['ID'])]['Rating'] = 'No Rating'
        else:
            song2['song{0}'.format(row['ID'])]['Rating'] = float(row['Rating'])
        s = row['Song'].split(' (')
        s = s[0].split(' feat.')
        s = s[0].split(' - ')
        song2['song{0}'.format(row['ID'])]['Song'] = s[0]
        song2['song{0}'.format(row['ID'])]['Artist 1'] = row['Artist 1']
        song2['song{0}'.format(row['ID'])]['Artist 2'] = row['Artist 2']
        if row['Year'] == '':
            song2['song{0}'.format(row['ID'])]['Re_Date'] = 'Unknown'
        else:
            song2['song{0}'.format(row['ID'])]['Re_Date'] = row['Year']
# print(song2)
```

```python
df = pd.read_csv(New)
df = df.copy()

F = 0
NR = 0
CL = 0

for y in range(1, len(song1)+1): # No_Dups
    s1 = song1['song{0}'.format(y)]['Song']
    a1 = song1['song{0}'.format(y)]['Artist 1']
    for i in range (1, len(song2)+1): # MusicColl
        s2 = song2['song{0}'.format(i)]['Song']
        a21 = song2['song{0}'.format(i)]['Artist 1']
        a22 = song2['song{0}'.format(i)]['Artist 2']

        if (s1 == s2 and a1 == a21) or (s1 == s2 and a1 == a22):
            df.loc[y-1, 'Re_Date'] = song2['song{0}'.format(i)]['Re_Date']
            if df.loc[y-1, 'Rating'] != '-':
                df.loc[y-1, 'Rating'] = song2['song{0}'.format(i)]['Rating']
                F = F + 1
            else:
                NR = NR + 1
            break
        # elif s1 == s2:
            # print(y, s1, a1)
            # print(i, s2, a21, a22)
            # print()
        elif i == len(song2):
            df.loc[y-1, 'Rating'] = 'Can\'t Locate'
            df.loc[y-1, 'Re_Date'] = 'Unknown'
            CL = CL + 1
            # print(s1, '\n')

    if df.loc[y-1, 'Spot_Date'] != 'Unknown' and df.loc[y-1, 'Re_Date'] != 'Unknown':
        print(int(df.loc[y-1, 'Spot_Date'][:4]))
        print(int(df.loc[y-1, 'Re_Date']))
        if int(df.loc[y-1, 'Spot_Date'][:4]) <= int(df.loc[y-1, 'Re_Date']):
            df.loc[y-1, 'Date'] = df.loc[y-1, 'Spot_Date'][:4]
        elif int(df.loc[y-1, 'Spot_Date'][:4]) >= int(df.loc[y-1, 'Re_Date']):
            df.loc[y-1, 'Date'] = df.loc[y-1, 'Re_Date']
    elif df.loc[y-1, 'Spot_Date'] != 'Unknown':
        df.loc[y-1, 'Date'] = df.loc[y-1, 'Spot_Date'][:4]
    elif df.loc[y-1, 'Re_Date'] != 'Unknown':
        df.loc[y-1, 'Date'] = df.loc[y-1, 'Re_Date']
    else:
        df.loc[y-1, 'Date'] = 'Unknown'

# Implement Changes to csv
df.to_csv(r'C:\Users\mackt\Python\Music Library\Data\Music_Library.csv', index =False)
```
